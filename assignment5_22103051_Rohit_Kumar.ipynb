```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Question 1(a):- Read the csv file. Find out the number of classes of cylinders (cyl) present in the dataset.


```python
car_info = pd.read_csv("car_info.csv")  # here we read the data 
car_info
```

```python
car_info1= car_info.drop_duplicates(['cyl']) #here we remove all the duplicates of 
no_of_class = car_info1['cyl']
print(no_of_class)
print("number of differet classes in cylinder=",len(no_of_class)) 
```

Question 1(b):- Plot a scatter plot of “mpg” vs “hp”. Use different point types for different class of
cylinder. Provide axis title, plot title, legend appropriately. 

```python
sns.scatterplot(x=car_info['hp'],y=car_info['mpg'],hue = car_info['cyl'], s=100,style= car_info['cyl'], palette="viridis")
plt.show()
```

Question 1(c):- Now use different sub plots to plot “mpg” vs “hp” for different classes of cylinders. Each sub plot should be for each unique value of cylinders. Provide all axis titles, plot title, sub plot title appropriately. 

```python
plt.figure(figsize=(10,15)) 
ax= plt.subplot(3,1,1)
sns.scatterplot(x=car_info[car_info['cyl']==4]['hp'],y=car_info[car_info['cyl']==4]['mpg'], data=car_info) 
plt.xlabel('hp for class=4')
ax= plt.subplot(3,1,2)
sns.scatterplot(x=car_info[car_info['cyl']==6]['hp'],y=car_info[car_info['cyl']==6]['mpg'], data=car_info) 
plt.xlabel('hp for class=6')
ax= plt.subplot(3,1,3)
sns.scatterplot(x=car_info[car_info['cyl']==8]['hp'],y=car_info[car_info['cyl']==8]['mpg'], data=car_info) 
plt.xlabel('hp for class=8')
plt.tight_layout() 
```
